```python
import time
from collections import defaultdict
import copy
import emojis

Same = list()
x = 0
emoji_dic = {
    1 : emojis.encode(":sunglasses:"),
    2 : emojis.encode(":joy:"),
    3 : emojis.encode(":heart_eyes:"),
    4 : emojis.encode(":open_mouth:"),
    5 : emojis.encode(":sleeping:"),
    6 : emojis.encode(":confused:")
}
Different = []
IS = []
Pre = []
num = []
CS = []
num_state = []
goal_state = []
total_time = 0
number_of_nodes = 0
c = 0
t = time.time()

n = int(input("Enter the order of puzzle"))
for r in range(n):
    num_state.append([0] * n)
    Pre.append([0] * n)
    for c in range(n):
        num_state[r][c] = int(input())
        if(num_state[r][c] == 1):
            Pre[r][c] = emoji_dic[1]
        elif(num_state[r][c] == 2):
            Pre[r][c] = emoji_dic[2] 
        elif(num_state[r][c] == 3):
            Pre[r][c] = emoji_dic[3]
        elif(num_state[r][c] == 4):
            Pre[r][c] = emoji_dic[4]
        elif(num_state[r][c] == 5):
            Pre[r][c] = emoji_dic[5]
        elif(num_state[r][c] == 6):
            Pre[r][c] = emoji_dic[6]

print("\n")
print("Initial State :")
for i in range(n):
    print(Pre[i])
list1 = list()
list1.append(Pre)
count = defaultdict(int)
for i in Pre:
    for j in i:
        count[j] += 1
#print(count)
list2 = list()
list2.append(Pre)


def Check(CS):
    for r in range(n):
        for c in range(n):
            if r >= 0 and c == 0:
                if CS[r][c] == CS[r][c + 1]:
                    return False
                if (r < n - 1):
                    if CS[r][c] == CS[r + 1][c]:
                        return False
                if (r > 0):
                    if CS[r][c] == CS[r - 1][c]:
                        return False

            if r >= 0 and c == n - 1:
                if CS[r][c] == CS[r][c - 1]:
                    return False

                if (r < n - 1):
                    if CS[r][c] == CS[r + 1][c]:
                        return False

                if (r > 0):
                    if CS[r][c] == CS[r - 1][c]:
                        return False

            if c >= 0 and r == 0:
                if CS[r][c] == CS[r + 1][c]:
                    return False

                if (c < n - 1):
                    if CS[r][c] == CS[r][c + 1]:
                        return False

                if (c > 0):
                    if CS[r][c] == CS[r][c - 1]:
                        return False

            if c >= 0 and r == n - 1:
                if CS[r][c] == CS[r - 1][c]:
                    return False

                if (c < n - 1):
                    if CS[r][c] == CS[r][c + 1]:
                        return False

                if (c > 0):
                    if CS[r][c] == CS[r][c - 1]:
                        return False

            if c > 0 and c < n - 1 and r > 0 and r < n - 1:
                if CS[r][c] == CS[r + 1][c]:
                    return False

                if CS[r][c] == CS[r - 1][c]:
                    return False

                if CS[r][c] == CS[r][c + 1]:
                    return False

                if CS[r][c] == CS[r][c - 1]:
                    return False

    return True


while len(list1) != 0:
    IS = list1.pop(0)

    #    if Check(IS)==True:
    #        print(IS)

    for r in range(0, n):
        for c in range(0, n):
            if r >= 0 and c == 0:
                if IS[r][c] == IS[r][c + 1]:
                    if (r, c + 1) not in Same:
                        Same.append((r, c + 1))
                        x = x + 1

                else:
                    if (r, c + 1) not in Different:
                        Different.append((r, c + 1))
                        x = x + 1

                if (r < n - 1):
                    if IS[r][c] == IS[r + 1][c]:
                        if (r + 1, c) not in Same:
                            Same.append((r + 1, c))
                            x = x + 1
                    else:
                        if (r + 1, c) not in Different:
                            Different.append((r + 1, c))
                            x = x + 1

                if (r > 0):
                    if IS[r][c] == IS[r - 1][c]:
                        if (r - 1, c) not in Same:
                            Same.append((r - 1, c))
                            x = x + 1
                    else:
                        if (r - 1, c) not in Different:
                            Different.append((r - 1, c))
                            x = x + 1
            if r >= 0 and c == n - 1:
                if IS[r][c] == IS[r][c - 1]:
                    if (r, c - 1) not in Same:
                        Same.append((r, c - 1))
                        x = x + 1
                else:
                    if (r, c - 1) not in Different:
                        Different.append((r, c - 1))
                        x = x + 1

                if (r < n - 1):
                    if IS[r][c] == IS[r + 1][c]:
                        if (r + 1, c) not in Same:
                            Same.append((r + 1, c))
                            x = x + 1
                    else:
                        if (r + 1, c) not in Different:
                            Different.append((r + 1, c))
                            x = x + 1

                if (r > 0):
                    if IS[r][c] == IS[r - 1][c]:
                        if (r - 1, c) not in Same:
                            Same.append((r - 1, c))

                    else:
                        if (r - 1, c) not in Different:
                            Different.append((r - 1, c))
                            x = x + 1

            if c >= 0 and r == 0:
                if IS[r][c] == IS[r + 1][c]:
                    if (r + 1, c) not in Same:
                        Same.append((r + 1, c))
                        x = x + 1
                else:
                    if (r + 1, c) not in Different:
                        Different.append((r + 1, c))
                        x = x + 1

                if (c < n - 1):
                    if IS[r][c] == IS[r][c + 1]:
                        if (r, c + 1) not in Same:
                            Same.append((r, c + 1))
                            x = x + 1
                    else:
                        if (r, c + 1) not in Different:
                            Different.append((r, c + 1))
                            x = x + 1

                if (c > 0):
                    if IS[r][c] == IS[r][c - 1]:
                        if (r, c - 1) not in Same:
                            Same.append((r, c - 1))
                            x = x + 1
                    else:
                        if (r, c - 1) not in Different:
                            Different.append((r, c - 1))
                            x = x + 1

            if c >= 0 and r == n - 1:
                if IS[r][c] == IS[r - 1][c]:
                    if (r - 1, c) not in Same:
                        Same.append((r - 1, c))
                        x = x + 1
                else:
                    if (r - 1, c) not in Different:
                        Different.append((r - 1, c))
                        x = x + 1

                if (c < n - 1):
                    if IS[r][c] == IS[r][c + 1]:
                        if (r, c + 1) not in Same:
                            Same.append((r, c + 1))
                            x = x + 1
                    else:
                        if (r, c + 1) not in Different:
                            Different.append((r, c + 1))
                            x = x + 1

                if (c > 0):
                    if IS[r][c] == IS[r][c - 1]:
                        if (r, c - 1) not in Same:
                            Same.append((r, c - 1))
                            x = x + 1
                    else:
                        if (r, c - 1) not in Different:
                            Different.append((r, c - 1))
                            x = x + 1

            if c > 0 and c < n - 1 and r > 0 and r < n - 1:
                if IS[r][c] == IS[r + 1][c]:
                    if (r + 1, c) not in Same:
                        Same.append((r + 1, c))
                        x = x + 1
                else:
                    if (r + 1, c) not in Different:
                        Different.append((r + 1, c))
                        x = x + 1

                if IS[r][c] == IS[r - 1][c]:
                    if (r - 1, c) not in Same:
                        Same.append((r - 1, c))
                        x = x + 1
                else:
                    if (r - 1, c) not in Different:
                        Different.append((r - 1, c))
                        x = x + 1

                if IS[r][c] == IS[r][c + 1]:
                    if (r, c + 1) not in Same:
                        Same.append((r, c + 1))
                        x = x + 1
                else:
                    if (r, c + 1) not in Different:
                        Different.append((r, c + 1))
                        x = x + 1
                if IS[r][c] == IS[r][c - 1]:
                    if (r, c - 1) not in Same:
                        Same.append((r, c - 1))
                        x = x + 1
                else:
                    if (r, c - 1) not in Different:
                        Different.append((r, c - 1))
                        x = x + 1
            
            if Same != 0 and Different != 0:
                while len(Different) != 0:
                    CS = copy.deepcopy(IS)
                    p, q = Different.pop(0)
                    CS[r][c], CS[p][q] = CS[p][q], CS[r][c]
                    # print(CS)
                    if Check(CS) == True:
                        goal_state = CS
                        q = time.time()
                        total_time = q-t
                        number_of_nodes = x
                        exit()

                    else:

                        if CS not in list2:
                            list2.append(CS)
                            if CS in list2:
                                list1.append(CS)



print("\nGoal State :")
for i in range(n):
    print(goal_state[i])

print("\nTime Required :", total_time)
print("Number of nodes visited in BFS traversal :", number_of_nodes, '\n')


                                
# print(Same)
# print("hello")
# print(Different)


```
